```python
# Program menampilkan nama sebanyak 200 kali
for i in range(1, 201):
    print(f"{i}. Andre")
```

```python
# Program menampilkan angka 1 sampai 500
for i in range(1, 501):
    print(i)
```

```python
# Program menampilkan bilangan ganjil dari 1 sampai 200
i = 1
while i <= 200:
    print(i)
    i += 2
```

```python
# Menggunakan for loop
jumlah_genap = 0
for i in range(1, 101):
    if i % 2 == 0:
        jumlah_genap += i
print("Jumlah bilangan genap (for loop):", jumlah_genap)

# Menggunakan while loop
i = 1
jumlah_genap_while = 0
while i <= 100:
    if i % 2 == 0:
        jumlah_genap_while += i
    i += 1
print("Jumlah bilangan genap (while loop):", jumlah_genap_while)
```
